```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

moca_cn = df_metadata[df_metadata["Group"] == "Control"]["Lenght (s)"].dropna()
moca_ad = df_metadata[df_metadata["Group"] == "Patient"]["Lenght (s)"].dropna()

print("Control:", moca_cn.min(), moca_cn.max())
print("Patient:", moca_ad.min(), moca_ad.max())

min_val = min(moca_cn.min(), moca_ad.min())
max_val = max(moca_cn.max(), moca_ad.max())

bins = np.linspace(min_val, max_val, 20)

palette = {
    "Control": sns.color_palette("Set2")[0],
    "Patient": sns.color_palette("Set2")[1]
}

fig, axes = plt.subplots(
    2, 1,
    figsize=(12, 8),
    sharex=True,
    sharey=True
)

axes[0].hist(
    moca_cn,
    bins=bins,
    color=palette["Control"],
    edgecolor="black",
    alpha=0.75
)

axes[0].set_title("Control")
axes[0].set_ylabel("Número de audios")

axes[1].hist(
    moca_ad,
    bins=bins,
    color=palette["Patient"],
    edgecolor="black",
    alpha=0.75
)

axes[1].set_title("Patient")
axes[1].set_xlabel("Duración (s)")
axes[1].set_ylabel("Número de audios")

fig.suptitle("Audio duration distribution by group")

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df_plot = df.copy()

df_plot['label'] = df_plot['label'].replace({
    0: 'CN',
    1: 'AD',
    '0': 'CN',
    '1': 'AD'
})

cols_features = [
    c for c in df_plot.columns
    if isinstance(c, str) and (c.startswith('score') or c.startswith('count_')) and c != 'score_total'
]

df_long = df_plot.melt(
    id_vars='label',
    value_vars=cols_features,
    var_name='feature',
    value_name='valor'
)

df_long = df_long.dropna(subset=['valor'])

g = sns.FacetGrid(
    df_long,
    col="feature",
    hue="label",
    col_wrap=3,
    height=4,
    sharex=False,
    sharey=False,
    palette="Set2"
)

g.map(
    sns.histplot,
    "valor",
    discrete=True,
    multiple="dodge",
    shrink=0.8
)

g.add_legend(title="Clase")

g.set_axis_labels("Puntuación", "Número de muestras")
g.set_titles("{col_name}")

for ax in g.axes.flat:
    ax.grid(axis="y", linestyle="--", alpha=0.4)

plt.subplots_adjust(top=0.9)
g.fig.suptitle("Score distribution by task and class")

plt.show()
```
